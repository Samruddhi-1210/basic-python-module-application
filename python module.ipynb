```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.bar(df["Name"],df["percentage"])
plt.title("Student Percentage Graph")
plt.xlabel("student_name")
plt.ylabel("percentage")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(df["Name"],df["percentage"],color="lavender")
plt.title("Student Percentage Graph")
plt.xlabel("student_name")
plt.ylabel("percentage")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Name"],df["percentage"],marker="o")
plt.title("Student Percentage Graph")
plt.xlabel("student_name")
plt.ylabel("percentage")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(df["Attendance"],labels=df["Name"],autopct="%1.1f%%")
plt.title("Student Percentage Graph")
plt.xlabel("student_name")
plt.ylabel("percentage")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df["Maths"],df["Science"])
plt.title("Student maths vs Science Percentage Graph")
plt.xlabel("Math")
plt.ylabel("science")
plt.grid()
plt.show()
```
